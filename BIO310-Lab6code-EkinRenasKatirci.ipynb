```python
def rc(seq):
    """
    Function to reverse complement a DNA string.
    e.g.
    "AGGAATGAAC" -> "GTTCATTCCT"

    :param seq: A string of DNA consisting of characters A C T G

    :return rc_seq:  is a string of DNA consisting of characters A C T G
    """

    # Your code here
    comp_dict = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    rc_seq = "".join(comp_dict[base] for base in seq[::-1])

    return rc_seq



def getkmers(seq, k):
    """
    Function to obtain list of kmers from a sequence.

    :param seq: A string of DNA consisting of characters A C T G; e.g. 'AGGAATGAAC'

    :param k: is an int e.g. k = 4

    :return kmers: is a list of kmers e.g. ['AGGA', 'GGAA', 'GAAT', 'AATG', 'ATGA', 'TGAA', 'GAAC']
    """

    # Your code here
    kmers = []
    for i in range(len(seq) - k + 1):
        kmers.append(seq[i:i+k])

    return kmers



def DeBruijn(kmers):
    """
    :param kmers: kmers is a list such as generated from getkmers(seq, k).
    e.g.
    ['AGGA', 'GGAA', 'GAAT', 'AATG', 'ATGA', 'TGAA', 'GAAC']

    :return deBruijn: is a dict of the deBruijn graph
    e.g.
    {'GAA': ['AAC', 'AAT'],
    'AGG': ['GGA'],
    'GGA': ['GAA'],
    'AAT': ['ATG'],
    'TGA': ['GAA'],
    'ATG': ['TGA']}
    """

    # Your code here
    deBruijn = {}
    for kmer in kmers:
      prefix = kmer[:-1]
      suffix = kmer[1:]
      if prefix in deBruijn:
            deBruijn[prefix].append(suffix)
      else:
          deBruijn[prefix] = [suffix]

    return deBruijn
```

```python
##########################################################
#### DO NOT MODIFY BELOW #################################
# After completing above functions and uploading the reads.txt to colab runtime, run this cell. #######
##########################################################


def genomeAssembly(dna):
  l = len(dna[0]) # Assuming all reads are same length
  for k in range(l-1, 1, -1):
      deBruijn_per_seq = [DeBruijn(getkmers(seq, k)) for seq in dna]
      deBruijn_genome = {key:val for d in deBruijn_per_seq for key,val in d.items()}

      first = kmer = next(iter(deBruijn_genome))
      superstring = ''

      while True:
        if kmer in deBruijn_genome:
          superstring += kmer[-1]
          kmer = deBruijn_genome.pop(kmer)[0]
          if kmer == first:
            return superstring, k
        else:
          break

with open('/content/reads_lab6.txt', 'r') as infile:
    dna = infile.read().strip().split('\n')
    dna = list(set(dna + [rc(i) for i in dna]))

# You will report the genome, and the integer k that is used for k-mers to assemble genome, and the length of the genome.
genome, k = genomeAssembly(dna)
genome, k, len(genome)
```
